```python
import pandas as pd
df = pd.read_csv("youtubers_2.csv", encoding = "utf8")
df
```

```python
import ast  # ast模組可以將字符串解析為Python對象

data = []

for index, row in df.iterrows():
    name = row['Youtuber頻道名稱']
    # 使用ast.literal_eval將字符串解析為列表
    try:
        video_links = ast.literal_eval(row['所有影片連結'])
        if isinstance(video_links, list):
            for link in video_links:
                data.append({'youtuber_channel': name, 'video_link': link})
    except ValueError:
        # 如果row['所有影片連結']不是一個可解析的列表字符串，將會在這裡捕捉到ValueError
        print(f"Error parsing video links for {name}")

# 用pd.concat來建立DataFrame
video_df = pd.concat([pd.DataFrame([i]) for i in data], ignore_index=True)
        
video_df
```

```python
video_df.to_csv("video_links_2.csv", encoding = "utf-8", index = False)
```
