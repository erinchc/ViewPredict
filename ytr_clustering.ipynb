```python
import pandas as pd
data = pd.read_csv("youtubers.csv", encoding = "utf8")
data = data[['Youtuber頻道名稱','開始經營時間','總訂閱數(萬)']]
data
```

```python
import datetime as dt
data['開始經營時間'] = pd.to_datetime(data['開始經營時間'])
today = pd.Timestamp('now')
data['channel_age_days'] = (today - data['開始經營時間']).dt.days
data = data.drop(['開始經營時間'], axis = 1)
data = data.rename(columns = {'Youtuber頻道名稱':'channel_name', '總訂閱數(萬)':'subscribers'})
data
```

```python
def convert_unit(x):
    return int(float(x)*10000)

data['subscribers'] = data['subscribers'].apply(convert_unit)
data.head()
```

```python
data.set_index('channel_name', inplace=True)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
    
# Plotting the Elbow Curve to determine the optimal number of clusters
import matplotlib.pyplot as plt
plt.plot(range(1,11), inertia, marker = 'o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()
```

```python
ideal_k = 3
kmeans = KMeans(n_clusters = ideal_k)
kmeans.fit(scaled_data)
data['Cluster'] = (kmeans.labels_ +1)
data
```
