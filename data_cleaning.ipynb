```python
df['publish_date'] = pd.to_datetime(df['publish_date'], errors='coerce', utc=True)

# 移除時區信息
df['publish_date'] = df['publish_date'].dt.tz_convert(None)

# 只保留日期部分
df['publish_date'] = df['publish_date'].dt.date

df.head()
```

```python
def remove_words(x):
    if "觀看次數：" in x:
        x = x.replace("觀看次數：","")
        return x

df['views'] = df['views'].apply(remove_words)
df
```

```python
data = df[df['views'].isna() == True]
data
```

```python
index_to_drop = [2401, 2402, 3650]
df = df.drop(index_to_drop)
```

```python
def convert_to_int(a):
    if "萬" in a:
        a = a.replace("萬次", "")
        return int(float(a)*10000)
    else:
        a = a.replace("次", "")
        return int(a)

df['views'] = df['views'].apply(convert_to_int)
df
```

```python
def replace(i):
    if "萬" in i:
        i = i.replace("萬","")
        return int(float(i)*10000)
    elif "喜歡" in i:
        return 0
    else:
        return int(i)

df['likes'] = df['likes'].apply(replace)
df[2000:2050]
```

```python
def remove_comma(y):
    if "," in y:
        y = y.replace(",", "")
        return int(y)
    else: 
        return int(y)

df["comments"] = df["comments"].apply(remove_comma)
df
```

```python
df.loc[390:569, "channel_name"] = "STR Network"
df.loc[570, "channel_name"] = "黃阿瑪的後宮生活"
df[389:571]
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
special_indices = df[df['channel_name'].str.startswith('@') | df['channel_name'].str.startswith('#')].index

# Replacing the special channel names with the channel name from the previous row
for index in special_indices:
    if index > 0:  # Ensuring we don't go out of bounds for the first row
        df.at[index, 'channel_name'] = df.at[index - 1, 'channel_name']

# Display the updated data for the previously identified special cases
updated_special_rows = df.loc[special_indices]
print(updated_special_rows.head())
```

```python
missing_channel_names = df[df['channel_name'].isnull() | (df['channel_name'] == '')]

for index in missing_channel_names.index:
    # Check the previous row if not the first row
    if index > 0 and pd.notna(df.at[index - 1, 'channel_name']):
        df.at[index, 'channel_name'] = df.at[index - 1, 'channel_name']
    # Else, check the next row if not the last row
    elif index < (len(df) - 1) and pd.notna(df.at[index + 1, 'channel_name']):
        df.at[index, 'channel_name'] = df.at[index + 1, 'channel_name']

# Displaying the updated data for the previously identified missing 'channel_name' cases
updated_missing_channel_names = df.loc[missing_channel_names.index]
updated_missing_channel_names.head()
```
