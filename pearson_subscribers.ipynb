```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np

df1 = pd.read_csv('youtubers.csv', encoding = 'utf8')
```

```python
views = df1['總觀看數(萬)']
subscribers = df1['總訂閱數(萬)']
correlation = views.corr(subscribers)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=views, y=subscribers)
plt.title(f'views vs subscribers (Pearson Correlation: {correlation:.2f})')
plt.xlabel('Views (10K)')
plt.ylabel('Subscribers (10K))')
plt.grid(True)
plt.show()
```
